````python
import time
from IPython.display import clear_output, display

def jupyter_timer(seconds):
    """Displays a live countdown timer in a Jupyter Notebook cell."""
    for remaining in range(seconds, -1, -1):
        # Clear the output of the current cell before rendering new text
        clear_output(wait=True)
        
        # Display the formatted countdown string
        display(f"Time remaining: {remaining} seconds")
        
        if remaining > 0:
            time.sleep(1)
            
    print("Time is up!")

# Example: Start a 10-second timer
# jupyter_timer(10)
````
