```python
#WHILE
#The while loop exists as a tool for repeated execution based on a condition
#The code block will keep being executed until the given logical condition returns a False boolean value

#iterate through list dates and stop at the year 1973, then print out the number of iterations.

dates = [1982, 1980, 1973, 2000]

i=0
year=dates[0]

while(year!=1973):
    print(year)
    i=i+1
    year=dates[i]

    print("It took",i,"repetations to get out of loop.")
```

```python
#Controlling Loop Execution with break and continue

#The break statement

for num in range(1, 10):
     if num == 5:
        print("Breaking the loop at:", num)
        break
print(num)
```

```python
#The continue statement
#continue and skip just the num**

for num in range(1, 6):
    if num == 2:
        continue
    print(num)
```

```python
#Using break and continue in a while loop

count = 0
while count < 10:
    count += 1
    if count == 3:
        continue  # skip printing 3
    if count == 8:
        break     # stop the loop when count is 8
    print(count)
```

```python
#Write a while loop to copy the strings 'orange' of the list squares to the list new_squares. 
#Stop and exit the loop if the value on the list is not 'orange':

squares = ['orange', 'orange', 'purple', 'blue ', 'orange']

new_squares = []
i=0

while (i<len(squares) and squares[i]=="orange"):
    new_squares.append(squares[i])
    i=i+1
    print(new_squares)
```

```python
#Write a Python program using a for loop that prints numbers from 1 to 15 but:
#Skips multiples of 3
#Stops the loop if the number is greater than 12

for i in range(1,16):
    if i%3==0: # skip multiples of 3
        continue
    if i>12:
        break  # stop if number > 12
    print(i)
```
