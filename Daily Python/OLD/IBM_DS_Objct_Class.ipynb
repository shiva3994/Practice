```python
# Import the library

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a class Circle

class Circle(object):
    
    # Constructor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius                         #Define your class
        self.color = color 
    
    # Method
    def add_radius(self, r):
        self.radius = self.radius + r                #Data attributes used to initialize object
        return(self.radius)
    
    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')                           #Method used to r to radious
        plt.show() 
```

```python
# Create an object RedCircle

RedCircle = Circle(10, 'red')
```

```python
# Find out the methods can be used on the object RedCircle

dir(RedCircle)
```

```python
# Create a new Rectangle class for creating a rectangle object

class Rectangle(object):
    
    # Constructor
    def __init__(self, width=2, height=3, color='r'):
        self.height = height 
        self.width = width
        self.color = color
    
    # Method
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()
        
```

```python
# Create a new object rectangle
BlueRectangle = Rectangle(10,20,"Blue")
BlueRectangle.height
BlueRectangle.width
BlueRectangle.color
BlueRectangle.drawRectangle()
```

```python
#Create yellow rectangle
YellowRectangle = Rectangle(25,20,"Yellow")
YellowRectangle.height
YellowRectangle.width
YellowRectangle.color
YellowRectangle.drawRectangle()
```
