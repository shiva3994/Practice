```python
# If statement example *****************
age1 = 20

#expression that can be true or false
if age1 > 18:
    
    #within an indent, we have the expression that is run if the condition is true
    print("Accepted above 18")

#The statements after the if statement will run regardless if the condition is true or false 
print("Visit cartoon centre")
```

```python
#My age
age2 = 20

#If age below 20 then appear NDA.
if age2 < 20:

    #If variable age is below 20 this condition will be true.
    print("Below 20 - Appear NDA")

#If value of age variable is not below 20 - above condtition will be false and this condition will run regardless.
print("Above 20 - Try CDS")
```

```python
# Else statement example *****************

#My age
age3 = 30

#If age below 22 then appear CDS.
if age3 < 24:
    
    #If age is below 24 this condtion will be true and run
    print("You can appear for CDS")

else:
    #If age is not below 24 and above condition is false then this condtion will be true and run
    print("You can appear for TA")

#If value of age variable is true or false above condtition will run regardless.
print("Move on JAI HIND")
```

```python
# Elif statment example *****************

#My age
age4 = 32

#If age below or equal to 20 then condition meet and opertaion halts here and goes to last operation ** print("Best of Luck !") **
if age4 <= 20:

    print("You can appear for NDA")

#If age above 20 1st contion not meet - operation will come to condition 2. If age below 20 and equal 25 this operation will run else go to ** ELIF ** 
elif age4 <= 25:

    print("You can appear for CDS")

#if age above 20 and 25 both conditions not met so operation will come to ELIF and last operation will rul along with last operation
else:

    print("You can appear for TA")
    
print("Best of Luck !")
```

```python
# While Loop Example
dates = [1982, 1980, 2000,1973]

i = 0
year = dates[0]

while(year != 1973):    
    print(year)
    i = i + 1
    year = dates[i]
    
print("It took", i ,"repetitions to get out of loop.")
```

```python
#The break statement
#It immediately terminates the loop it appears in, regardless of the loop’s condition. 
#It’s commonly used when a specific condition is met and no further iterations are needed.
for num in range(1, 10):
    if num == 5:
        print("Breaking the loop at:", num)
        break
    print(num)
```

```python
#The continue statement
#It continue statement skips the current iteration and moves to the next one in the loop. 
#It’s used when you want to ignore certain conditions but continue looping.
for num in range(1, 6):
    if num == 3:
        continue
    print(num)
```

```python
#Using break and continue in a while loop
count = 0
while count < 10:
    count += 1
    if count == 3:
        continue  # skip printing 3
    if count == 8:
        break     # stop the loop when count is 8
    print(count)
```
